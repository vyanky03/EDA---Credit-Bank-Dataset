## Univariate Analysis

### Numerical Columns

```python
m_num_col = ['AMT_CREDIT','AMT_APPLICATION']
```

```python
for col in m_num_col:
    sns.distplot(x=merge_data[col])
    plt.title(f'{col}')
    plt.xticks(rotation=90)
    plt.show()
```

### Conclusion-

### Categorical Columns

```python
m_cat_col =['NAME_CONTRACT_TYPE','CODE_GENDER','NAME_INCOME_TYPE','OCCUPATION_TYPE','AMT_CREDIT_GRADE','NAME_CONTRACT_STATUS','AMT_APPLICATION_GRADE']
```

```python
for col in m_cat_col:
    sns.countplot(x=merge_data[col],data=merge_data)
    plt.xticks(rotation=90)
    plt.show()
```

### Conclusion-

- Loans taken by female customers is more than male customers
- Most of the customers who has taken loans has income type as worker
- Most of the customers who has taken loans has occupation type as laborers
- Approved loans has higher number than cancelled or refused loans
